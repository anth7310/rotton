```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('starwars_2116.csv', header=0)
df.head()
```

```python
# cleaning
features = ['review', 'displayName', 'isVerified', 'isSuperReviewer', 'hasSpoilers', 
 'hasProfanity', 'createDate', 'updateDate', 'score']
df[features].head()
```

```python
df[features].describe(include="all")
```

```python
df = df[features]
# change boolean values to binary
bf = ['isVerified', 'isSuperReviewer', 'hasSpoilers', 'hasProfanity']
df[bf] = (df[bf] == 'True').astype(int)
df.head()
```

```python
df.describe(include="all")
```

```python
df['review_length'] = df.apply(lambda x : len(x['review']), axis=1)
df.head()
```
